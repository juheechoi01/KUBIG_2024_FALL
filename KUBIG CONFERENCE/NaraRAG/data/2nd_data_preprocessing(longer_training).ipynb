```python
df['question'] = df['question'].apply(lambda x: eval(x))
df['answer'] = df['answer'].apply(lambda x: eval(x))

df['question'][0]['comment']
```

```python
df['question_comment'] = df['question'].apply(lambda x: x['comment'])
df['question_keyword'] = df['question'].apply(lambda x: x['keyword'])
df['answer_comment'] = df['answer'].apply(lambda x: x['comment'])
df['answer_keyword'] = df['answer'].apply(lambda x: x['keyword'])
```

```python
questioner = []
answerer = []
for idx, row in df.iterrows():
    try:
        q_name = row['questioner_name']+' '+row['questioner_affiliation']+' '+row['questioner_position']
    except:
        if type(row['questioner_affiliation'])==float:
            if type(row['questioner_position'])==float:
                q_name = row['questioner_name']
            else:
                q_name = row['questioner_name']+' '+row['questioner_position']
        else:
            q_name = row['questioner_name']+' '+row['questioner_affiliation']
            
    
    try:
        a_name = row['answerer_name']+' '+row['answerer_affiliation']+' '+row['answerer_position']
    except:
        if type(row['answerer_affiliation'])==float:
            if type(row['answerer_position'])==float:
                a_name = row['answerer_name']
            else: a_name = row['answerer_name']+' '+row['answerer_position']
        else: a_name = row['answerer_name']+' '+row['answerer_affiliation']
            
    
    questioner.append(q_name)
    answerer.append(a_name)
```
